```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model, optimizers
import umap
import warnings
warnings.filterwarnings('ignore')

# Set seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Set the style for plots
plt.style.use('fivethirtyeight')
sns.set_palette("viridis")
plt.rcParams['figure.figsize'] = (12, 8)

# Function to load and preprocess the data
def load_data(file_path='player_centric_data.csv'):
    """Load and preprocess the chess player data"""
    # Load the dataset
    df = pd.read_csv(file_path)
    return df

def get_features_and_types(df):
    """Extract features and player types from the dataframe"""
    # Define metadata columns to exclude from features
    metadata_cols = ['player_name', 'player_color', 'opponent_name', 'player_elo', 
                   'player_type', 'player_type_color', 'game_id', 'result']
    
    # Get numeric features
    numeric_features = [col for col in df.columns 
                      if col not in metadata_cols and pd.api.types.is_numeric_dtype(df[col])]
    
    # Get player types for labeled data
    player_types = df['player_type'].copy()
    
    # Create a clean dataset with imputed missing values
    imputer = SimpleImputer(strategy='median')
    X = pd.DataFrame(imputer.fit_transform(df[numeric_features]), columns=numeric_features)
    
    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    return X_scaled, player_types, numeric_features

def visualize_latent_space(X_2d, player_types=None, method_name=''):
    """Visualize the latent space in 2D with optional player type coloring"""
    plt.figure(figsize=(14, 10))
    
    # If player types are available, color by type
    if player_types is not None:
        # Remove NaN values
        mask = ~player_types.isna()
        X_2d_labeled = X_2d[mask]
        types_labeled = player_types[mask].values
        
        unique_types = np.unique(types_labeled)
        colors = plt.cm.tab10(np.linspace(0, 1, len(unique_types)))
        
        for i, t in enumerate(unique_types):
            mask_t = (types_labeled == t)
            plt.scatter(X_2d_labeled[mask_t, 0], X_2d_labeled[mask_t, 1], 
                        c=[colors[i]], label=t, alpha=0.6, s=30)
        
        plt.legend()
    else:
        plt.scatter(X_2d[:, 0], X_2d[:, 1], alpha=0.6, s=30)
    
    plt.title(f'{method_name} Latent Space')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    plt.tight_layout()
    plt.savefig(f'{method_name.lower().replace(" ", "_")}_latent_space.png', dpi=300)
    plt.show()

def evaluate_clusters(X_2d, player_types, method_name=''):
    """Cluster the 2D latent space and evaluate against player types"""
    # Filter to only include labeled data points
    mask = ~player_types.isna()
    X_2d_labeled = X_2d[mask]
    types_labeled = player_types[mask].reset_index(drop=True)
    
    # Apply K-means clustering
    n_clusters = len(types_labeled.unique())
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_2d_labeled)
    
    # Create contingency table
    contingency = pd.crosstab(types_labeled, labels)
    
    # Calculate metrics
    ari = adjusted_rand_score(types_labeled, labels)
    nmi = normalized_mutual_info_score(types_labeled, labels)
    
    # Calculate cluster purity
    purity = 0
    for i in range(contingency.shape[1]):
        purity += contingency.iloc[:, i].max()
    purity /= contingency.values.sum()
    
    print(f"\n{method_name} Clustering Evaluation:")
    print(f"Adjusted Rand Index: {ari:.4f}")
    print(f"Normalized Mutual Information: {nmi:.4f}")
    print(f"Cluster Purity: {purity:.4f}")
    
    print("\nContingency Table:")
    print(contingency)
    
    # Visualize clusters vs actual types
    plt.figure(figsize=(20, 10))
    
    # Plot clusters
    plt.subplot(1, 2, 1)
    scatter = plt.scatter(X_2d_labeled[:, 0], X_2d_labeled[:, 1], c=labels, cmap='viridis', alpha=0.6, s=30)
    plt.colorbar(scatter, label='Cluster')
    plt.title(f'{method_name} K-means Clusters (k={n_clusters})')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    
    # Plot player types
    plt.subplot(1, 2, 2)
    
    # Encode player types numerically for coloring
    le = LabelEncoder()
    type_codes = le.fit_transform(types_labeled)
    
    scatter = plt.scatter(X_2d_labeled[:, 0], X_2d_labeled[:, 1], c=type_codes, cmap='tab10', alpha=0.6, s=30)
    plt.colorbar(scatter, label='Player Type')
    plt.title('Actual Player Types')
    plt.xlabel('Component 1')
    plt.ylabel('Component 2')
    
    # Add a legend
    handles = []
    for i, t in enumerate(le.classes_):
        handles.append(plt.Line2D([0], [0], marker='o', color='w', 
                                markerfacecolor=plt.cm.tab10(i/len(le.classes_)), 
                                markersize=10, label=t))
    plt.legend(handles=handles)
    
    plt.tight_layout()
    plt.savefig(f'{method_name.lower().replace(" ", "_")}_clusters.png', dpi=300)
    plt.show()
    
    return {
        'method': method_name,
        'ari': ari,
        'nmi': nmi,
        'purity': purity,
        'contingency': contingency
    }

#------------------------------------------------------------
# 1. AUTOENCODER
#------------------------------------------------------------
def build_autoencoder(input_dim, encoding_dim=2, hidden_layers=[32, 16, 8]):
    """Build an autoencoder model with the specified architecture"""
    # Input layer
    input_layer = keras.Input(shape=(input_dim,))
    
    # Encoder
    encoded = input_layer
    for units in hidden_layers:
        encoded = layers.Dense(units, activation='relu')(encoded)
    
    # Bottleneck layer (latent space)
    bottleneck = layers.Dense(encoding_dim, name='bottleneck')(encoded)
    
    # Decoder (mirror of encoder)
    decoded = bottleneck
    for units in reversed(hidden_layers):
        decoded = layers.Dense(units, activation='relu')(decoded)
    
    # Output layer
    output_layer = layers.Dense(input_dim)(decoded)
    
    # Create model
    autoencoder = Model(input_layer, output_layer)
    encoder = Model(input_layer, bottleneck)
    
    # Compile model
    autoencoder.compile(optimizer='adam', loss='mse')
    
    return autoencoder, encoder

def train_autoencoder(X, encoding_dim=2, epochs=100, batch_size=32, validation_split=0.2):
    """Train an autoencoder and return the encoder model"""
    print("\n=== AUTOENCODER ===")
    
    # Build the autoencoder
    autoencoder, encoder = build_autoencoder(X.shape[1], encoding_dim)
    
    # Print model summary
    print("\nAutoencoder model summary:")
    autoencoder.summary()
    
    # Train the model
    history = autoencoder.fit(
        X, X,
        epochs=epochs,
        batch_size=batch_size,
        shuffle=True,
        validation_split=validation_split,
        verbose=1
    )
    
    # Plot training history
    plt.figure(figsize=(12, 5))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Autoencoder Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.savefig('autoencoder_training_history.png', dpi=300)
    plt.show()
    
    # Get the latent space representation
    X_encoded = encoder.predict(X)
    
    return X_encoded, encoder, autoencoder

#------------------------------------------------------------
# 2. VARIATIONAL AUTOENCODER (VAE)
#------------------------------------------------------------
def build_vae(input_dim, latent_dim=2, hidden_layers=[32, 16, 8]):
    """Build a Variational Autoencoder (VAE) model"""
    # Input layer
    inputs = keras.Input(shape=(input_dim,))
    
    # Encoder
    x = inputs
    for units in hidden_layers:
        x = layers.Dense(units, activation='relu')(x)
    
    # VAE latent space parameters
    z_mean = layers.Dense(latent_dim, name='z_mean')(x)
    z_log_var = layers.Dense(latent_dim, name='z_log_var')(x)
    
    # Sampling layer
    def sampling(args):
        z_mean, z_log_var = args
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.random.normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon
    
    z = layers.Lambda(sampling, output_shape=(latent_dim,), name='z')([z_mean, z_log_var])
    
    # Build encoder model
    encoder = Model(inputs, [z_mean, z_log_var, z], name='encoder')
    
    # Build decoder model
    latent_inputs = keras.Input(shape=(latent_dim,), name='z_sampling')
    x = latent_inputs
    for units in reversed(hidden_layers):
        x = layers.Dense(units, activation='relu')(x)
    outputs = layers.Dense(input_dim)(x)
    decoder = Model(latent_inputs, outputs, name='decoder')
    
    # VAE model
    outputs = decoder(encoder(inputs)[2])
    vae = Model(inputs, outputs, name='vae')
    
    # Add KL divergence regularization
    reconstruction_loss = tf.reduce_mean(tf.reduce_sum(tf.keras.losses.mse(inputs, outputs), axis=1))
    kl_loss = -0.5 * tf.reduce_mean(1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
    vae_loss = reconstruction_loss + kl_loss
    
    vae.add_loss(vae_loss)
    vae.compile(optimizer='adam')
    
    return vae, encoder, decoder

def train_vae(X, latent_dim=2, epochs=100, batch_size=32, validation_split=0.2):
    """Train a Variational Autoencoder and return the encoder model"""
    print("\n=== VARIATIONAL AUTOENCODER (VAE) ===")
    
    # Build the VAE
    vae, encoder, decoder = build_vae(X.shape[1], latent_dim)
    
    # Print model summary
    print("\nVAE model summary:")
    vae.summary()
    
    # Train the model
    history = vae.fit(
        X, None,  # VAE calculates its own loss
        epochs=epochs,
        batch_size=batch_size,
        shuffle=True,
        validation_split=validation_split,
        verbose=1
    )
    
    # Plot training history
    plt.figure(figsize=(12, 5))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('VAE Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.savefig('vae_training_history.png', dpi=300)
    plt.show()
    
    # Get the latent space representation
    z_mean, z_log_var, z = encoder.predict(X)
    
    return z_mean, encoder, vae

#------------------------------------------------------------
# 3. DEEP EMBEDDING CLUSTERING (DEC)
#------------------------------------------------------------
def build_deep_embedding_clustering(input_dim, n_clusters, hidden_layers=[32, 16, 8], latent_dim=10):
    """Build a Deep Embedding Clustering (DEC) model"""
    # Input layer
    inputs = keras.Input(shape=(input_dim,))
    
    # Encoder
    x = inputs
    for units in hidden_layers:
        x = layers.Dense(units, activation='relu')(x)
    
    # Latent space
    embedding = layers.Dense(latent_dim, name='embedding')(x)
    
    # Build encoder model
    encoder = Model(inputs, embedding, name='encoder')
    
    # Clustering layer
    clustering_layer = ClusteringLayer(n_clusters, name='clustering')(embedding)
    
    # Build DEC model
    dec_model = Model(inputs, [embedding, clustering_layer], name='dec')
    
    return dec_model, encoder

# Custom Clustering Layer for DEC
class ClusteringLayer(layers.Layer):
    def __init__(self, n_clusters, weights=None, alpha=1.0, **kwargs):
        super(ClusteringLayer, self).__init__(**kwargs)
        self.n_clusters = n_clusters
        self.alpha = alpha
        self.initial_weights = weights
        self.input_spec = layers.InputSpec(ndim=2)
        
    def build(self, input_shape):
        assert len(input_shape) == 2
        input_dim = input_shape[1]
        
        self.clusters = self.add_weight(
            shape=(self.n_clusters, input_dim),
            initializer='glorot_uniform',
            name='clusters'
        )
        
        if self.initial_weights is not None:
            self.set_weights(self.initial_weights)
            del self.initial_weights
        self.built = True
        
    def call(self, inputs):
        q = 1.0 / (1.0 + tf.reduce_sum(
            tf.square(tf.expand_dims(inputs, axis=1) - self.clusters), axis=2) / self.alpha)
        q = q ** ((self.alpha + 1.0) / 2.0)
        q = tf.transpose(tf.transpose(q) / tf.reduce_sum(q, axis=1))
        return q
    
    def compute_output_shape(self, input_shape):
        assert input_shape and len(input_shape) == 2
        return (input_shape[0], self.n_clusters)
    
    def get_config(self):
        config = {'n_clusters': self.n_clusters, 'alpha': self.alpha}
        base_config = super(ClusteringLayer, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))

def train_dec(X, player_types, latent_dim=10, epochs=100, batch_size=32, n_clusters=4):
    """Train a Deep Embedding Clustering model"""
    print("\n=== DEEP EMBEDDING CLUSTERING (DEC) ===")
    
    # First train a simple autoencoder for pre-training
    autoencoder, encoder = build_autoencoder(X.shape[1], latent_dim)
    
    autoencoder.compile(optimizer='adam', loss='mse')
    autoencoder.fit(X, X, batch_size=batch_size, epochs=50, verbose=1)
    
    # Use the encoder to get embeddings
    embeddings = encoder.predict(X)
    
    # Initialize cluster centers using K-means
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)
    kmeans.fit(embeddings)
    cluster_centers = kmeans.cluster_centers_
    
    # Build DEC model
    dec_model, dec_encoder = build_deep_embedding_clustering(
        X.shape[1], n_clusters, latent_dim=latent_dim)
    
    # Set encoder weights from autoencoder
    for i, layer in enumerate(dec_encoder.layers[1:]):  # Skip input layer
        layer.set_weights(encoder.layers[i+1].get_weights())
    
    # Set clustering layer weights from KMeans
    clustering_layer = dec_model.get_layer(name='clustering')
    clustering_layer.set_weights([cluster_centers])
    
    # Compile DEC model
    dec_model.compile(optimizer='adam', loss=['mse', 'kld'])
    
    # Generate target distribution
    def target_distribution(q):
        weight = q ** 2 / q.sum(0)
        return (weight.T / weight.sum(1)).T
    
    # Train DEC
    for i in range(epochs):
        if i % 10 == 0:
            embeddings, q = dec_model.predict(X)
            p = target_distribution(q)
            y_pred = q.argmax(1)
            
            # Calculate clustering metrics if player types available
            if player_types is not None:
                mask = ~player_types.isna()
                y_true = player_types[mask].values
                y_pred_masked = y_pred[mask]
                
                ari = adjusted_rand_score(y_true, y_pred_masked)
                nmi = normalized_mutual_info_score(y_true, y_pred_masked)
                
                print(f'Epoch {i}: ARI = {ari:.4f}, NMI = {nmi:.4f}')
        
        # Train on batch
        dec_model.fit(X, [X, p], batch_size=batch_size, epochs=1, verbose=0)
    
    # Get final embeddings and cluster assignments
    embeddings, q = dec_model.predict(X)
    cluster_assignments = q.argmax(1)
    
    # Reduce to 2D for visualization using UMAP
    umap_reducer = umap.UMAP(n_components=2, random_state=42)
    embeddings_2d = umap_reducer.fit_transform(embeddings)
    
    return embeddings_2d, cluster_assignments, dec_model

#------------------------------------------------------------
# 4. SIAMESE NETWORK
#------------------------------------------------------------
def build_siamese_network(input_dim, embedding_dim=16):
    """Build a Siamese Network for learning player similarity"""
    # Base network
    base_input = keras.Input(shape=(input_dim,))
    x = layers.Dense(64, activation='relu')(base_input)
    x = layers.Dense(32, activation='relu')(x)
    x = layers.Dense(embedding_dim, activation=None)(x)
    embedding = layers.Lambda(lambda x: tf.math.l2_normalize(x, axis=1))(x)
    base_network = Model(base_input, embedding, name='base_network')
    
    # Siamese configuration
    input_a = keras.Input(shape=(input_dim,))
    input_b = keras.Input(shape=(input_dim,))
    
    embedding_a = base_network(input_a)
    embedding_b = base_network(input_b)
    
    # Calculate cosine similarity
    similarity = layers.Dot(axes=1, normalize=True)([embedding_a, embedding_b])
    
    # Create models
    siamese_model = Model(inputs=[input_a, input_b], outputs=similarity)
    embedding_model = Model(inputs=base_input, outputs=embedding)
    
    # Compile with contrastive loss
    siamese_model.compile(loss='binary_crossentropy', optimizer='adam')
    
    return siamese_model, embedding_model

def generate_triplets(X, player_types, num_triplets=10000):
    """Generate triplets (anchor, positive, negative) for triplet loss training"""
    triplets = []
    
    # Get player types that are not NaN
    mask = ~player_types.isna()
    X_labeled = X[mask]
    types_labeled = player_types[mask].reset_index(drop=True)
    
    # Get indices for each player type
    type_indices = {}
    for t in types_labeled.unique():
        type_indices[t] = np.where(types_labeled == t)[0]
    
    # Generate random triplets
    for _ in range(num_triplets):
        # Pick a random player type
        anchor_type = np.random.choice(list(type_indices.keys()))
        
        # Pick anchor and positive of same type
        anchor_idx = np.random.choice(type_indices[anchor_type])
        positive_idx = np.random.choice(type_indices[anchor_type])
        
        # Make sure anchor and positive are different
        while positive_idx == anchor_idx:
            positive_idx = np.random.choice(type_indices[anchor_type])
        
        # Pick a different player type for negative
        negative_type = np.random.choice(
            [t for t in type_indices.keys() if t != anchor_type])
        negative_idx = np.random.choice(type_indices[negative_type])
        
        triplets.append((anchor_idx, positive_idx, negative_idx))
    
    return X_labeled, types_labeled, triplets

def generate_pairs(X, player_types, num_pairs=10000):
    """Generate pairs for siamese network training"""
    pairs = []
    labels = []
    
    # Get player types that are not NaN
    mask = ~player_types.isna()
    X_labeled = X[mask]
    types_labeled = player_types[mask].reset_index(drop=True)
    
    # Get indices for each player type
    type_indices = {}
    for t in types_labeled.unique():
        type_indices[t] = np.where(types_labeled == t)[0]
    
    # Generate random pairs
    for _ in range(num_pairs):
        # Same class pairs (50% of pairs)
        if np.random.random() > 0.5:
            # Pick a random player type
            pair_type = np.random.choice(list(type_indices.keys()))
            
            # Pick two players of same type
            idx1 = np.random.choice(type_indices[pair_type])
            idx2 = np.random.choice(type_indices[pair_type])
            
            # Make sure samples are different
            while idx2 == idx1:
                idx2 = np.random.choice(type_indices[pair_type])
            
            pairs.append((idx1, idx2))
            labels.append(1)  # Same class
            
        # Different class pairs (50% of pairs)
        else:
            # Pick two different player types
            type1, type2 = np.random.choice(
                list(type_indices.keys()), size=2, replace=False)
            
            # Pick one player from each type
            idx1 = np.random.choice(type_indices[type1])
            idx2 = np.random.choice(type_indices[type2])
            
            pairs.append((idx1, idx2))
            labels.append(0)  # Different class
    
    return X_labeled, types_labeled, pairs, np.array(labels)

def train_siamese_network(X, player_types, embedding_dim=16, epochs=50, batch_size=128):
    """Train a Siamese Network for player similarity"""
    print("\n=== SIAMESE NETWORK ===")
    
    # Generate pairs
    X_labeled, types_labeled, pairs, pair_labels = generate_pairs(X, player_types)
    
    # Build siamese network
    siamese_model, embedding_model = build_siamese_network(X.shape[1], embedding_dim)
    
    # Print model summary
    print("\nSiamese model summary:")
    siamese_model.summary()
    
    # Create training data
    pairs_a = np.array([X_labeled[i] for i, _ in pairs])
    pairs_b = np.array([X_labeled[j] for _, j in pairs])
    
    # Train model
    history = siamese_model.fit(
        [pairs_a, pairs_b], pair_labels,
        validation_split=0.2,
        batch_size=batch_size,
        epochs=epochs,
        verbose=1
    )
    
    # Plot training history
    plt.figure(figsize=(12, 5))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Siamese Network Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.savefig('siamese_training_history.png', dpi=300)
    plt.show()
    
    # Get embeddings for all data
    embeddings = embedding_model.predict(X)
    
    # Reduce to 2D for visualization using UMAP
    umap_reducer = umap.UMAP(n_components=2, random_state=42)
    embeddings_2d = umap_reducer.fit_transform(embeddings)
    
    return embeddings_2d, embedding_model

#------------------------------------------------------------
# 5. ADVERSARIAL AUTOENCODER (AAE)
#------------------------------------------------------------
def build_adversarial_autoencoder(input_dim, latent_dim=2, hidden_layers=[64, 32]):
    """Build an Adversarial Autoencoder model"""
    # Encoder
    encoder_inputs = keras.Input(shape=(input_dim,))
    x = encoder_inputs
    for units in hidden_layers:
        x = layers.Dense(units, activation='relu')(x)
    encoded = layers.Dense(latent_dim, name='latent_space')(x)
    encoder = Model(encoder_inputs, encoded, name='encoder')
    
    # Decoder
    latent_inputs = keras.Input(shape=(latent_dim,))
    x = latent_inputs
    for units in reversed(hidden_layers):
        x = layers.Dense(units, activation='relu')(x)
    decoder_outputs = layers.Dense(input_dim)(x)
    decoder = Model(latent_inputs, decoder_outputs, name='decoder')
    
    # Discriminator
    discriminator_inputs = keras.Input(shape=(latent_dim,))
    x = discriminator_inputs
    x = layers.Dense(32, activation='relu')(x)
    x = layers.Dense(16, activation='relu')(x)
    discriminator_outputs = layers.Dense(1, activation='sigmoid')(x)
    discriminator = Model(discriminator_inputs, discriminator_outputs, name='discriminator')
    
    # Compile discriminator
    discriminator.compile(optimizer='adam', loss='binary_crossentropy')
    
    # Full Adversarial Autoencoder
    # Autoencoder part
    autoencoder_outputs = decoder(encoder(encoder_inputs))
    autoencoder = Model(encoder_inputs, autoencoder_outputs, name='autoencoder')
    autoencoder.compile(optimizer='adam', loss='mse')
    
    # Generator-Discriminator part (GAN)
    discriminator.trainable = False
    gan_outputs = discriminator(encoder(encoder_inputs))
    gan = Model(encoder_inputs, gan_outputs, name='gan')
    gan.compile(optimizer='adam', loss='binary_crossentropy')
    
    return encoder, decoder, discriminator, autoencoder, gan

def train_adversarial_autoencoder(X, latent_dim=2, epochs=100, batch_size=128):
    """Train an Adversarial Autoencoder model"""
    print("\n=== ADVERSARIAL AUTOENCODER (AAE) ===")
    
    # Build the AAE
    encoder, decoder, discriminator, autoencoder, gan = build_adversarial_autoencoder(X.shape[1], latent_dim)
    
    # Print model summaries
    print("\nEncoder model summary:")
    encoder.summary()
    
    print("\nDecoder model summary:")
    decoder.summary()
    
    print("\nDiscriminator model summary:")
    discriminator.summary()
    
    # Training loop
    autoencoder_losses = []
    discriminator_losses = []
    generator_losses = []
    
    n_samples = X.shape[0]
    
    for epoch in range(epochs):
        # Sample minibatch
        idx = np.random.randint(0, n_samples, batch_size)
        real_batch = X[idx]
        
        # Train autoencoder
        autoencoder_loss = autoencoder.train_on_batch(real_batch, real_batch)
        autoencoder_losses.append(autoencoder_loss)
        
        # Train discriminator
        # Real samples (from normal distribution)
        real_z = np.random.normal(size=(batch_size, latent_dim))
        discriminator_real = discriminator.train_on_batch(real_z, np.ones((batch_size, 1)))
        
        # Fake samples (from encoder)
        fake_z = encoder.predict(real_batch)
        discriminator_fake = discriminator.train_on_batch(fake_z, np.zeros((batch_size, 1)))
        
        discriminator_loss = 0.5 * (discriminator_real + discriminator_fake)
        discriminator_losses.append(discriminator_loss)
        
        # Train generator (encoder) to fool discriminator
        generator_loss = gan.train_on_batch(real_batch, np.ones((batch_size, 1)))
        generator_losses.append(generator_loss)
        
        # Print progress
        if epoch % 10 == 0:
            print(f"Epoch {epoch}/{epochs}")
            print(f"Autoencoder Loss: {autoencoder_loss:.4f}")
            print(f"Discriminator Loss: {discriminator_loss:.4f}")
            print(f"Generator Loss: {generator_loss:.4f}")
    
    # Plot training progress
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    plt.plot(autoencoder_losses)
    plt.title('Autoencoder Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    
    plt.subplot(1, 3, 2)
    plt.plot(discriminator_losses)
    plt.title('Discriminator Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    
    plt.subplot(1, 3, 3)
    plt.plot(generator_losses)
    plt.title('Generator Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    
    plt.tight_layout()
    plt.savefig('aae_training_history.png', dpi=300)
    plt.show()
    
    # Get the latent space representation
    X_encoded = encoder.predict(X)
    
    return X_encoded, encoder

#------------------------------------------------------------
# 6. SELF-SUPERVISED LEARNING WITH CONTRASTIVE LOSS
#------------------------------------------------------------
def build_contrastive_model(input_dim, embedding_dim=16, temperature=0.5):
    """Build a model for self-supervised contrastive learning"""
    # Base encoder
    inputs = keras.Input(shape=(input_dim,))
    x = layers.Dense(128, activation='relu')(inputs)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dense(32, activation='relu')(x)
    embeddings = layers.Dense(embedding_dim)(x)
    embeddings = layers.Lambda(lambda x: tf.math.l2_normalize(x, axis=1))(embeddings)
    
    encoder = Model(inputs, embeddings, name='encoder')
    
    # Contrastive model
    anchor = keras.Input(shape=(input_dim,))
    positive = keras.Input(shape=(input_dim,))
    
    # Get embeddings for anchor and positive
    anchor_embedding = encoder(anchor)
    positive_embedding = encoder(positive)
    
    # Create models
    contrastive_model = Model(inputs=[anchor, positive], 
                             outputs=[anchor_embedding, positive_embedding])
    
    # Define contrastive loss function
    def contrastive_loss(y_true, y_pred):
        # Extract anchor and positive embeddings
        anchor_emb, positive_emb = y_pred
        
        # Calculate similarity matrix
        similarity = tf.matmul(anchor_emb, positive_emb, transpose_b=True)
        
        # Apply temperature scaling
        similarity /= temperature
        
        # Create labels (diagonal ones)
        batch_size = tf.shape(anchor_emb)[0]
        labels = tf.eye(batch_size)
        
        # Calculate cross-entropy loss
        loss = tf.keras.losses.categorical_crossentropy(
            y_true=labels, y_pred=tf.nn.softmax(similarity), from_logits=False)
        
        return tf.reduce_mean(loss)
    
    # Compile model with custom loss
    contrastive_model.compile(optimizer='adam', loss=contrastive_loss)
    
    return contrastive_model, encoder

def generate_augmented_pairs(X, n_pairs=1000, noise_level=0.05):
    """Generate augmented pairs for contrastive learning"""
    pairs = []
    
    # Generate random indices
    indices = np.random.choice(len(X), size=n_pairs)
    
    for idx in indices:
        # Original sample
        anchor = X[idx]
        
        # Create augmentation by adding random noise
        noise = np.random.normal(0, noise_level, size=anchor.shape)
        positive = anchor + noise
        
        pairs.append((anchor, positive))
    
    return np.array([p[0] for p in pairs]), np.array([p[1] for p in pairs])

def train_contrastive_model(X, embedding_dim=16, epochs=50, batch_size=128):
    """Train a self-supervised model with contrastive learning"""
    print("\n=== SELF-SUPERVISED CONTRASTIVE LEARNING ===")
    
    # Build model
    contrastive_model, encoder = build_contrastive_model(X.shape[1], embedding_dim)
    
    # Print model summary
    print("\nEncoder model summary:")
    encoder.summary()
    
    # Training loop
    for epoch in range(epochs):
        # Generate augmented pairs
        anchors, positives = generate_augmented_pairs(X, n_pairs=batch_size*10)
        
        # Train in batches
        n_batches = len(anchors) // batch_size
        epoch_loss = 0
        
        for batch in range(n_batches):
            start_idx = batch * batch_size
            end_idx = (batch + 1) * batch_size
            
            anchor_batch = anchors[start_idx:end_idx]
            positive_batch = positives[start_idx:end_idx]
            
            # Dummy labels (not used in custom loss)
            dummy_labels = np.zeros((batch_size, 1))
            
            # Train on batch
            loss = contrastive_model.train_on_batch(
                [anchor_batch, positive_batch], dummy_labels)
            
            epoch_loss += loss
        
        # Print progress
        if epoch % 5 == 0:
            print(f"Epoch {epoch}/{epochs}, Loss: {epoch_loss/n_batches:.4f}")
    
    # Get embeddings for all data
    embeddings = encoder.predict(X)
    
    # Reduce to 2D for visualization using UMAP
    umap_reducer = umap.UMAP(n_components=2, random_state=42)
    embeddings_2d = umap_reducer.fit_transform(embeddings)
    
    return embeddings_2d, encoder

#------------------------------------------------------------
# 7. COMPARE ALL NEURAL NETWORK METHODS
#------------------------------------------------------------
def run_all_neural_methods(X, player_types, epochs=50, batch_size=32):
    """Run all neural network methods and compare results"""
    print("\n=== RUNNING ALL NEURAL NETWORK METHODS ===")
    
    results = []
    
    # 1. Autoencoder
    print("\nTraining Autoencoder...")
    ae_latent, encoder, autoencoder = train_autoencoder(
        X, encoding_dim=2, epochs=epochs, batch_size=batch_size)
    
    visualize_latent_space(ae_latent, player_types, "Autoencoder")
    if player_types is not None:
        ae_results = evaluate_clusters(ae_latent, player_types, "Autoencoder")
        results.append(ae_results)
    
    # 2. VAE
    print("\nTraining Variational Autoencoder...")
    vae_latent, vae_encoder, vae = train_vae(
        X, latent_dim=2, epochs=epochs, batch_size=batch_size)
    
    visualize_latent_space(vae_latent, player_types, "Variational Autoencoder")
    if player_types is not None:
        vae_results = evaluate_clusters(vae_latent, player_types, "Variational Autoencoder")
        results.append(vae_results)
    
    try:
        # 3. AAE (Adversarial Autoencoder)
        print("\nTraining Adversarial Autoencoder...")
        aae_latent, aae_encoder = train_adversarial_autoencoder(
            X, latent_dim=2, epochs=epochs, batch_size=batch_size)
        
        visualize_latent_space(aae_latent, player_types, "Adversarial Autoencoder")
        if player_types is not None:
            aae_results = evaluate_clusters(aae_latent, player_types, "Adversarial Autoencoder")
            results.append(aae_results)
    except:
        print("Error training Adversarial Autoencoder. Skipping...")
    
    # 4. Siamese Network (only if player types available)
    if player_types is not None and (~player_types.isna()).sum() > 0:
        print("\nTraining Siamese Network...")
        siamese_latent, siamese_encoder = train_siamese_network(
            X, player_types, embedding_dim=16, epochs=epochs, batch_size=batch_size)
        
        visualize_latent_space(siamese_latent, player_types, "Siamese Network")
        siamese_results = evaluate_clusters(siamese_latent, player_types, "Siamese Network")
        results.append(siamese_results)
    
    # 5. Contrastive Learning
    print("\nTraining Contrastive Learning Model...")
    contrastive_latent, contrastive_encoder = train_contrastive_model(
        X, embedding_dim=16, epochs=epochs, batch_size=batch_size)
    
    visualize_latent_space(contrastive_latent, player_types, "Contrastive Learning")
    if player_types is not None:
        contrastive_results = evaluate_clusters(contrastive_latent, player_types, "Contrastive Learning")
        results.append(contrastive_results)
    
    try:
        # 6. DEC (Deep Embedding Clustering)
        if player_types is not None and (~player_types.isna()).sum() > 0:
            n_clusters = len(player_types.dropna().unique())
            print("\nTraining Deep Embedding Clustering...")
            dec_latent, dec_labels, dec_model = train_dec(
                X, player_types, latent_dim=10, epochs=epochs, batch_size=batch_size, n_clusters=n_clusters)
            
            visualize_latent_space(dec_latent, player_types, "Deep Embedding Clustering")
            dec_results = evaluate_clusters(dec_latent, player_types, "Deep Embedding Clustering")
            results.append(dec_results)
    except:
        print("Error training Deep Embedding Clustering. Skipping...")
    
    # Compare results
    if player_types is not None and results:
        print("\n=== NEURAL NETWORK METHODS COMPARISON ===")
        comparison_df = pd.DataFrame([{
            'Method': r['method'],
            'ARI': r['ari'],
            'NMI': r['nmi'],
            'Purity': r['purity']
        } for r in results])
        
        comparison_df = comparison_df.sort_values('NMI', ascending=False)
        print(comparison_df)
        
        # Plot comparison
        plt.figure(figsize=(14, 8))
        
        ax1 = plt.subplot(1, 1, 1)
        x = np.arange(len(comparison_df))
        width = 0.3
        
        ax1.bar(x - width, comparison_df['ARI'], width, label='ARI')
        ax1.bar(x, comparison_df['NMI'], width, label='NMI')
        ax1.bar(x + width, comparison_df['Purity'], width, label='Purity')
        
        ax1.set_xticks(x)
        ax1.set_xticklabels(comparison_df['Method'], rotation=45, ha='right')
        ax1.set_ylabel('Score')
        ax1.set_title('Neural Network Method Comparison')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('neural_network_comparison.png', dpi=300)
        plt.show()
        
        return comparison_df
    
    return None

# Main function to run all experiments
if __name__ == "__main__":
    # Load data
    df = load_data()
    
    # Extract features and player types
    X, player_types, feature_names = get_features_and_types(df)
    
    # Run all neural network methods
    comparison = run_all_neural_methods(X, player_types, epochs=30, batch_size=64)
    
    print("Neural network analysis complete!")
```
